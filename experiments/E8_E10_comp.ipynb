```python
import json
import matplotlib.pyplot as plt

# load 4 json files of score trajectories
# Load E8 trajectories
with open('../e8a_bscore_traj_se.json', 'r') as f:
    e8_se_traj = json.load(f)

with open('../e8a_bscore_traj_rb.json', 'r') as f:
    e8_rb_traj = json.load(f)

# Load E10 trajectories
with open('../e10_bscore_traj_se.json', 'r') as f:
    e10_se_traj = json.load(f)

with open('../e10_bscore_traj_rb.json', 'r') as f:
    e10_rb_traj = json.load(f)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 3))

# SummEval subplot
ax1.grid(True, which='both', linestyle='--', linewidth=0.5)
ax1.plot(e8_se_traj, color='blue', label='E8 SummEval')
ax1.plot(e10_se_traj, color='red', label='E10 SummEval')
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.set_xlabel('Optimization step')
ax1.set_ylabel('Best score')
ax1.legend()
ax1.set_title('SummEval')

# RewardBench subplot
ax2.grid(True, which='both', linestyle='--', linewidth=0.5)
ax2.plot(e8_rb_traj, color='blue', label='E8 RewardBench')
ax2.plot(e10_rb_traj, color='red', label='E10 RewardBench')
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.set_xlabel('Optimization step')
ax2.set_ylabel('Best score')
ax2.legend()
ax2.set_title('RewardBench')

plt.tight_layout()
plt.show()

# save figure
fig.savefig('e8_e10_comp.png', dpi=300)
```
